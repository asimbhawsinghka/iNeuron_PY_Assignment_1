### Solution 3:

```python
first_name = input("Please enter the first name:")
last_name = input("Please enter the last name:")

full_name = first_name + " " + last_name
full_name_reverse = ""

for i in range(1,len(full_name)+1,1):
  full_name_reverse = full_name_reverse + full_name[-i]

print("Full Name in reverse =", full_name_reverse)
```

### Solution 1:

```python
input_val = input("Enter a sequence of comma-seperated numbers: ")

output_val = list()

for i in input_val:
  if i == ",":
    continue
  output_val.append(int(i))

print("Output list = ",output_val)
```

### Solution 2:

```python
for i in range(1,10):
    if (i <= 5):
        num_stars = i
    else:
        num_stars = 10 - i
    for j in range(num_stars):
        print("*", end = "")
    print("")
```
